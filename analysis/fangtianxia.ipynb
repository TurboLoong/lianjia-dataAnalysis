```python
import pandas as pd
df = pd.read_csv('../house_data.csv', index_col=0)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import numpy as np
df.loc[df['物 业 费'] == '暂无资料', '物 业 费'] = np.nan
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.drop('参考月供', axis=1)
```

```python
df.head()
```

```python
df[(df['建筑面积']>100) & (df['總價'] > 2000) ].head() 
```

```python
df.rename(columns={'物 业 费': '物业费'}, inplace=True)
```

```python
df.head()
```

```python
df['物业费'] = df['物业费'].fillna(0)
```

```python
df['总价'].isnull().sum()
```

```python
df[df['总价'].isnull()]
```

```python
df['总价'].fillna(df['建筑面积']*(df['总价']/df['建筑面积']).mean(), inplace=True)
```

```python
df['总价'].isnull().sum()
```

```python
df.to_csv('../house_final.csv', index_label=False)
```
